```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Load the datasets
We have three files:
- `train.csv`: training dataset with survival labels
- `test.csv`: test dataset without survival labels
- `gender_submission.csv`: sample submission file

```python
gender_submission = pd.read_csv("gender_submission.csv")
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

train.shape, test.shape, gender_submission.shape
```

## Inspect the data
We’ll look at the first few rows, data types, and missing values.

```python
train.head()
train.info()
train.isnull().sum()
```

## Exploratory Data Analysis
We’ll explore survival counts, distributions, and relationships between features.

```python
# Survival counts
sns.countplot(x="Survived", data=train)
plt.title("Survival Counts")
plt.show()
```

```python
# Survival by Passenger Class
sns.countplot(x="Pclass", hue="Survived", data=train)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# Survival by Gender
sns.countplot(x="Sex", hue="Survived", data=train)
plt.title("Survival by Gender")
plt.show()
```

```python
# Age distribution
plt.hist(train["Age"].dropna(), bins=30, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Fare distribution
plt.hist(train["Fare"], bins=30, edgecolor="black")
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
# Survival by Embarked
sns.countplot(x="Embarked", hue="Survived", data=train)
plt.title("Survival by Embarkation Port")
plt.show()
```

## Correlation Analysis
We’ll check correlations between numerical features and survival.

## Feature Engineering
We’ll create a new feature: **FamilySize = SibSp + Parch + 1**.

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
sns.countplot(x="FamilySize", hue="Survived", data=train)
plt.title("Survival by Family Size")
plt.show()
```

## Key Insights
- Women and children had higher survival rates.
- Higher-class passengers (Pclass 1) survived more often.
- Larger families had mixed survival chances; very large families struggled.
- Age and Fare show skewed distributions; younger passengers had better odds.
